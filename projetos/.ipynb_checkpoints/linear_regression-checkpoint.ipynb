```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
url= "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/FuelConsumptionCo2.csv"
```

```python
data = pd.read_csv(url)
data
```

```python
data.describe()
```

```python
data.hist()
plt.tight_layout() #ajusta as distancias entre os hist
plt.show()
```

```python
plt.scatter(data.ENGINESIZE, data.FUELCONSUMPTION_CITY, color='red')
plt.xlabel("ENGINESIZE")
plt.ylabel("FUELCONSUMPTION_CITY")
plt.xlim(0,31)
plt.show
#aqui selecionamos algumas features e estamos olhando a relaÃ§Ã£o grafica entre elas, nesse caso observamos uma relaÃ§Ã£o linear entre elas
```

```python
X = data.ENGINESIZE.to_numpy()
y = data.FUELCONSUMPTION_CITY.to_numpy()
#estamos pegando os dados do dataframe do pandas e passando para um array em numpy
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
#dividindo em conjunto de treinamento e em conjunto de teste, usando apenas 20% em conjunto de teste
```

```python
type(X_train), np.shape(X_train), np.shape(y_train)  #checando o tipo de dado e avaliando a dimensÃ£o do vetor
```

```python
#agora entraremos na etapa da criaÃ§Ã£o do modelo de regressÃ£o linear
```

```python
from sklearn import linear_model
regressor = linear_model.LinearRegression()
regressor.fit(X_train.reshape(-1,1), y_train)
print(regressor.coef_)
print(regressor.intercept_)
```

```python
plt.scatter(X_train, y_train, color="red") #grafico de pontos
plt.plot(X_train, regressor.coef_ * X_train + regressor.intercept_, color="green") #faz uma reta
plt.xlabel("X_train")
plt.ylabel("y_train")
plt.xlim(0,31)
plt.show()
```
